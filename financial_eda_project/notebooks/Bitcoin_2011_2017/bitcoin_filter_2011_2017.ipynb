```python
import pandas as pd

# Cargar datos
file_path = "D:/Proyectos Personales ML/Entrenamientos ML/BITCOIN/Bitcoin History.csv"
df = pd.read_csv(file_path)

# Convertir fechas
df['Date'] = pd.to_datetime(df['Date'])

# Filtrar período
start_date = '2011-01-01'
end_date = '2017-12-31'
df_filtered = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

print(f"Registros originales: {len(df)}")
print(f"Registros filtrados: {len(df_filtered)}")

# Guardar resultados
output_file = "D:/Proyectos Personales ML/Entrenamientos ML/BITCOIN/Bitcoin_2011_2017.csv"
df_filtered.to_csv(output_file, index=False)
print(f"Datos guardados en {output_file}")

# Visualizar
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df_filtered['Date'], df_filtered['Price'])
plt.title('Bitcoin Price (2011-2017)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
